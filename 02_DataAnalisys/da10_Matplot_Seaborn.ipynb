### Matplot
    파이썬으로 데이터 분석 결과를 시각화 하는 데에는 Matplot 라이브러리를 가장 많이 사용
    Matplot은 파이썬에서 2d형태의 그래프, 이미지 등을 그릴 때 사용
    실제로 과학컴퓨팅 분야나 인공지능 연구분야에서도 많이 사용됨
    여기서 Matplot 여러가지 기능 중 기본적인 기능만 다뤄볼 예정
    
    Matplot 모듈 중에는 다양한 서브 모듈들이 있는데, 그 중에서 가장 많이 사용되는 서브 모듈이 pyplot.

```python
import matplotlib.pyplot as plt
```

### 1. Basic Plot

```python
# 1. plot() : 직선 혹은 꺾은 선을 그릴때 사용
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
# 직선은 ?
plt.plot([10,20,30,40])
plt.show()
'''
# 2 그래프에 옵션 추가
1) 선의 모양, 색상
2) x,y축의 최소, 최대값 지정
3) 그래프의 타이틀
'''

plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.title('Graph Test')
plt.show()
```

### 2. 겹쳐진 그래프 그리기

```python
import numpy as np
```

```python
t = np.arange(0,5,0.2) #0에서 5미만, 0.2간격
t
'''
이 값을 x, y값으로 활용
'''
plt.plot(t,t,'b--')
plt.show()
```

```python
# plt.plot(t,t,'b--', t,t**2,'gs')
plt.plot(t,t,'b--', t,t**2,'gs', t,t**3,'r^')
plt.show()
```

```python
# 겹쳐진 그래프 일때는 각 그래프에 대한 정보를 표시해야 알 수 있다.
plt.plot(t,t,'b--', label='A Group')
plt.plot(t,t**2,'gs', label='B Group')
plt.plot(t,t**3,'r^', label='C Group')

# 이 부분을 지정해야 label이 나타난다.
plt.legend()
plt.title('Wrapped Graph')

plt.xlabel('Year')
plt.ylabel('Growth Rate')
plt.grid()
plt.show()
```

### 파형모양의 그래프

```python
t = np.arange(0,12,0.01)
'''
기본으로 존재하는 figure가 아닌 새로운 figure를 생성하기 위해서
plt.figure()를 사용
'''
plt.figure(figsize=(8,6))

plt.plot(t, np.sin(t), 'r',lw=10, label='sin') # lw : Line Width
plt.plot(t, np.cos(t), 'g', label='cos')

plt.grid()
plt.legend()


plt.xlabel('Time')
plt.ylabel('Ampleitude')

plt.title('Graph Test2')
plt.show()
```

### SubPlot

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### 히스토그램
    hist() 함수를 사용
    데이터의 빈도에 따라 막대 그래프의 높이가 결정

```python
plt.hist([1,1,2,3,4,5,6,6,7,8,10])
plt.show()
```

```python
dice = []
for i in range(5) : 
    dice.append(np.random.randint(1,7))
print(dice)
```

```python
'''
bins
해당 막대 영역을 얼마나 채우는지를 결정하는 변수
이 값이 크면 영역이 더 촘촘하게 나눠짐
작으면 분포가 뭉뚱그려진다.
'''
plt.hist(dice,bins=6)
plt.show()
```

```python
# 이번에 주사위를 100000번 돌렸다고 가정.
dice = []
for i in range(100000) : 
    dice.append(np.random.randint(1,7))
#print(dice)

plt.hist(dice)
plt.show()
# 주사위 던지는 횟수를 늘릴수록 특정한 주사위 숫자가 나오는 횟수가 전체의 1/6 에 가까워진다 --> 큰 수의 법칙
```

```python
mu, sigma=100,15
x = mu+sigma *np.random.randn(10000)
x.shape
```

```python
plt.hist(x)
```

```python
plt.hist(x,bins=50,
        facecolor='g',
        density=1,
        alpha = 0.25
        )
plt.xlabel('smart')
plt.ylabel('Probability')
plt.title('Histogram of IQ')

plt.text(60,0.025,r'$mu=100, \sigma=15$') # r은 해석하지 말고 문자 그대로 받아들여라 : raw
plt.axis([42,160,0,0.028]) # x,y축의 보기 싫은 부분은 잘라낸다.
plt.show()
```

### Seaborn

```python
import seaborn as sns
import pandas as pd
# https://seaborn.pydata.org 갤러리 사이트
```

```python
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2000', periods=1000)) # 기본단위가 day
ts
```

```python
ts = ts.cumsum()
ts.plot() # x축으로는 2000/1/1 부터 1000일 기간, y축으로는 cumsum값
```

```python
ts.plot(kind='bar') # 시간이 꽤 걸림
```

### BoxPlot

```python
# 0~100까지 총 101개의 숫자를 만든다.
xs = np.array(np.linspace(start=0,stop=100, num=101))
df = pd.DataFrame(xs, columns=['feature'])
df
```

```python
plt.figure(figsize=(7,6)) # 크기지정
boxplot=df.boxplot(column=['feature'])
plt.yticks(np.arange(0,101,step=5))
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# total_bills값을 boxplot으로 표현
plt.figure(figsize=(6,3))
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
tips.sort_values(by='total_bill', ascending=False)
```

```python
# 날짜별로 데이터(total_bill)를 분석
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
'''
요일별, 흡연여부 별 전체음식값의 boxplot을 알아보자
 -> 어느 요일, 흡연자/비흡연자가 음식 지불액을 어떻게 지불했는지 확인
비교 -- hue
'''

sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
```

### LmPlot

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
'''
흡연자, 비흡연자 중에서 누가 더 팁을 많이 내는가
전체 음식지불액 함께 보면서. --> hue

음식값이 20달러를 넘어가는 순간부터 흡연자가 팁을 많이줌
'''

sns.lmplot(x='total_bill', y='tip',hue='smoker',data=tips)
```

### Heatmap

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights=flights.pivot(index='month',columns='year',values='passengers')
flights
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights) # Default : 값이 표현X
plt.show()
```

```python
'''
y축은 month
x축은 year
value는 승객수
'''
plt.figure(figsize=(10,8))
sns.heatmap(flights,annot=True, fmt="d") # 값 지정.
plt.show()
```
