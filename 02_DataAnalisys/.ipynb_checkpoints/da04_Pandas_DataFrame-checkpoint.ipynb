### DataFrame
#### dataFrame -생성
    DataFrame은 2차원 배열형식. 표같은 스프레드시트 자료구조
    여러개의 컬럼을 가지며 서로 다른 종류의 값이 담긴다.

    DataFrame은 다양한 방식으로 생성되지만
    가장 흔하게 생성하는 방법으로
    1)리스트를 값으로 가지는 딕셔너리 방법으로 생성
    2)Numpy 배열을 이용해서 생성
    3)read_csv(), read_excel()함수등을 이용해서 생성

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
# 1. 딕셔너리를 이용해서 생성
df_dic = {
        'state':['Ohio','Ohio', 'Ohio','Nevada','Nevada','Nevada'],
        'year': [2000,   2001,   2002,   2001,   2002,   2003],
         'pop': [ 1.5,   1.7,    3.6,    2.4,    2.9,    3.2]    
}

dicDf=DataFrame(df_dic)
dicDf
```

```python
print(type(dicDf.state))
print(type(dicDf.year))
# print(type(dicDf.pop))
print(type(dicDf['pop']))
```

```python
data1 = {'name':['James','Peter','Jane','Tomas'],
       'address':['NY','NY','LA','Texas'],
       'age':[33,44,55,66]      
      }
df1 = DataFrame(data1)
df1
```

```python
# 2. random 함수 사용하기
df2 = DataFrame(np.random.randint(10,100,16).reshape(4,4),
               columns=list('abcd'),
               index=list('abcd'))
# columns를 사용해서 컬럼명을 수정
df2.columns = ['one','two','three','four']
df2
```

```python
# 3. read_csv() 를 이용해서 생성
df3 =pd.read_csv('../data/tips.csv')
df3
```

#### DataFrame - 조회하기 : iloc, loc | iat, at
    iloc, loc를 주로 씀 (iloc : 인덱스, loc : 라벨)
    iat, at : 스칼라 값 조회

```python
# df2[df2['two']>45][['two']]
df2[df2['two']>45].iloc[:,1:2]
```

```python
# 1. James, Peter 라인 전부 가져오기...iloc, loc
print(df1.iloc[:2,])
print('*'*30)
print(df1.iloc[:2,:4])
print('*'*30)
print(df1.iloc[:2])

print('*'*30)
print('*'*30)
print(df1.loc[0:1, 'name':'age'])
```

```python
# 2. 스칼라값 가져오기 :: iat, at
print(df1.iat[1,1])
print(df1.iat[0,2])

print('*'*30)

print(df1.at[1,'address'])
print(df1.at[0,'age'])
```

```python
# 값 변경하기
df1.iat[2,2] = 50
df1
```

```python
# loc를 이용해서 마지막 라인에 NaN 추가하기
df1.loc[4] = np.nan
df1
```

```python
# 삭제하기 : drop 함수 -- 컬럼며을 삭제
df1.drop('phone',axis=1, inplace=True)
df1
```

```python
df1.drop(4,inplace=True)
df1
```

### Tips 실전 데이터로 응용하기

```python
# 1. 상위 3개의 데이터만 조회
df3.head(3)
```

```python
# # 2. tip비율로 정렬(높은 팁이 먼저나오도록), 샘플데이터 5개 출력

# df3.sort_values(by=['tip'], ascending=False).head(5)['tip'].plot(kind='bar')
# plt.show()

df3.sort_values(by=['tip'], ascending=False).head(5)['tip'].plot(kind='barh')
plt.show()
```
