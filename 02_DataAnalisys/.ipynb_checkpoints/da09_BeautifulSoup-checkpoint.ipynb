### 1. 네이버 URL 정보를 가져옴 : requests.get()
### 2. BeautifulSoup를 이용해서 파싱.

```python
import requests
from bs4 import BeautifulSoup

req = requests.get("https://www.naver.com") # 네이버 url 정보를 가져온다
html = req.text

soup = BeautifulSoup(html, 'html.parser') # BeautifulSoup를 이용해서 html.parser로 파싱.

# print(soup.prettify())
```

### 3. BeautifulSoup find() : 단수, find_all() : 복수

```python
result = soup.find_all('a','thumb') # a태그와 특정 클래스등을 참조한다.
# result

news_list = []
for i in result:
    news_list.append(i.find('img')['alt']) # img태그 안에 있는 alt속성의 값을 추가
print(news_list)
```

```python
# BeautifulSoup가 가지고 있는 특정 함수들
soup.title
soup.title.name
soup.title.string

# 태그가 여러개 있더라도 항상 첫번째 태그만 가져와서 파싱
soup.img
# soup.img['alt']
# sout.img['class'] class속성이 첫번째 img 태그에는 없기 때문에 에러난다
soup.img['height']

import re # 정규식 표현관련 모듈
print(soup.find_all(string=re.compile("네이버")))
```

### BeautifulSoup의 select_one(), select()
    fint(), find_all() -- 태그 기반 검색
    select_one(), select() -- 선택자 기반 검색

```python
print(soup.select_one('a')) # a태그 하나만 갖고옴
print(soup.select('a')) # a태그 다 갖고옴
```

```python
# css 선택자 사용하는 기법 그대로 사용
body = soup.select('body a')

# body
ul  = soup.select('div>ul')
ul
```

### BeautifulSoup get_Text(),get()
    get_text() --- 검색 결과에서 태그를 제외한 텍스트만 가져옴
    get('속성') --- 해당 속성의 값만 가져옴
    string() --- 검색결과에서 태그안의 또 다른 태그가 없는 경우에만 내용을 출력

```python
text = soup.find('span', attrs={'class':'blind'})
text
text.get_text()
text.get('class')
```

### BeautifulSoup를 활용해 네이버 영화 랭킹 가져오기
    1. 네이버 영화 랭킹 홈페이지(https://movie.naver.com/movie/sdb/rank/rmovie.nhn) 접속 : url, text 가져오기
    2. 분석 작업을 하기 위해서 BeautifulSoup을 만들고, html parser로 파싱
    3. 리턴된 soup 객체로 분석한 결과로 영화랭킹을 찾아서 출력
      -> select(), find_all()

```python
import requests
from bs4 import BeautifulSoup
```

```python
# 1. 특정 사이트 url을 받아옴
req = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.nhn')
html = req.text

# 2. 분석작업을 하기 위해서 bs 생성
soup = BeautifulSoup(html, 'html.parser')
```

```python
# 3. soup 객체를 분석해서 영화 랭킹을 찾아본다. (영화랭킹 태그가 어떤 구조로 되어있는가)
'''
<div class = 'tit3'>
    <a href>제목</a>
</div>
'''
movie_ranking_list = soup.find_all('div',class_='tit3')
# moive_ranking_list
for i in range(len(movie_ranking_list)):
    print("{:2}위 : {}".format(i+1,movie_ranking_list[i].get_text().strip()))
```
